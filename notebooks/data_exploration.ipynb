```python
import os
import matplotlib.pyplot as plt
import cv2
import random

# Set dataset path
dataset_path = "../dataset/"

# Get categories, excluding non-directory files
categories = [cat for cat in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, cat))]

# Visualize some sample images
fig, axes = plt.subplots(2, 3, figsize=(10, 7))

for i, category in enumerate(categories):
    category_path = os.path.join(dataset_path, category)
    if not os.listdir(category_path):  # Skip empty folders
        continue
    image_path = os.path.join(category_path, random.choice(os.listdir(category_path)))
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    axes[i//3, i%3].imshow(img)
    axes[i//3, i%3].set_title(category)
    axes[i//3, i%3].axis("off")

plt.tight_layout()
plt.show()
```
